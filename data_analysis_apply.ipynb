```python
import random

job_list = ["Teacher"] * 5 + ["Engineer"] *5 + ["RA"]*5 + ["Student"]*15
random.shuffle(job_list)

data = {
    "age" : range(20, 80, 2),
    "job": job_list
}


df = pd.DataFrame(data)
```

```python
df
```

```python
def old_young(row):
    
    if row['age'] <50:
        age_status = "junior"
        
    else:
        age_status = "senior"
        
    
    if row['job'] == 'Teacher' and age_status == "senior":
        return "Potential Retired Professor"
    
    else:
        return age_status + " " + row['job']


df["status"] = df.apply(old_young, axis=1)
```

```python
df
```

```python
for i, row in df.iterrows():
    age_value = df.loc[i, 'age']
    
    
    if age_value > 35:
        df.loc[i, 'age'] = 'Senior'
    else:
        df.loc[i, 'age'] = 'Junior'
        
```

```python
df.loc[0, 'status'] = 'changed'
```
