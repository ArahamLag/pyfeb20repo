```python
def generate_string(row):
    if row["A"] > 0:
        a_str = "P"
    else:
        a_str = "N"

        
    if row["B"] > 0:
        b_str = "P"
    else:
        b_str = "N"
    
    if row["C"] > 0:
        c_str = "P"
    else:
        c_str = "N"
    
    return a_str + b_str + c_str
```
